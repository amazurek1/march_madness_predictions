```python
def year_predictions (year1,year2):
    X = df[(year1<df['year']) & (df['year']<year2) | (df['year']>year2)].drop(['year','win'],axis=1)
    #could have also done answers1.set_index('SEQN') instead of dropping
    y = df[(year1<df['year']) & (df['year']<year2) | (df['year']>year2)]['win']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state = 123)
    tic = time()
    svclassifier = SVC(kernel='rbf', C=1000) 
    svclassifier.fit(X_train, y_train) 
    y_pred = svclassifier.predict(X_test)
    toc = time()
    print("run time is {} seconds".format(toc-tic))
    _2019_predict=svclassifier.predict(df[df['year']==year2].drop(['year','win'], axis = 1))
    print(confusion_matrix(df[df['year']==year2]['win'],_2019_predict))  
    print(classification_report(df[df['year']==year2]['win'],_2019_predict)) 
    print("The accuracy score is" + " "+ str(accuracy_score(df[df['year']==year2]['win'], _2019_predict)))
    running_list = []
    times_conflict = 0
    i = 0
    for num in list(_2019_predict):
        if i%2 ==0:
            running_list.append(num)
            i+=1
        else:
            if num == running_list[-1]:
                times_conflict+=1
                i+=1
            else:
                i+=1
                continue
    print(f'this is the times conflicted: {times_conflict}')



    
```
