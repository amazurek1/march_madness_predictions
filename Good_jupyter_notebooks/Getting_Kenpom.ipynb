```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as BS
```

```python
page = requests.get('https://kenpom.com/index.php?y=2019')
soup = BS(page.content, 'html.parser')
```

```python
idk = soup.select('tr')
```

```python
idk[2].select('a[href^="team.php?team="]')[0].string
```

```python
names = []
for thing in idk:
    try:
        names.append(thing.select('a[href^="team.php?team="]')[0].string)
    except:
        continue
```
